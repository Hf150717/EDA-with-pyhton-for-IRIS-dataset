```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import datasets
```

```python
df = pd.read_csv("IRIS.csv",sep=',',decimal='.')
df.head()
df.rename(columns={'sepal_length':'sepalUzunluk','sepal_width':'sepalGenişlik','petal_length':'petalUzunluk','petal_width':'petalGenişlik','species': 'tür'},inplace=True)
df.describe()
df.shape
df['tür'].value_counts()
```

```python
df["tür"]= df["tür"].replace("Iris-setosa",1) 
df["tür"]= df["tür"].replace("Iris-versicolor",2)
df["tür"]= df["tür"].replace("Iris-virginica", 3)
df.head(150)
```

```python
#Çizgi grafiği ile görselleştirme

plt.figure(figsize=(20,5))
for i,c in enumerate(['sepalUzunluk', 'sepalGenişlik', 'petalUzunluk', 'petalGenişlik']):
    plt.subplot(1,6,i+1)
    sns.distplot(df[df['tür'] == 1][c], kde=True, color='blue', hist=False, kde_kws=dict(linewidth=2), label=1)
    sns.distplot(df[df['tür'] == 2][c], kde=True, color='red', hist=False, kde_kws=dict(linewidth=2), label=2)
    sns.distplot(df[df['tür'] == 3][c], kde=True, color='green', hist=False, kde_kws=dict(linewidth=2), label=3)
    plt.title(c)
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
#burada kullanılmak istenen formüller csv formatında yükleme yapılan hale değil bu formata uygun olduğu için tekrar yükledik
iris = datasets.load_iris()
```

```python
#veri setinin mevcut  hali ve olması istenen halini atamak için değişkenler tanımladık
X = iris.data
y = iris.target
```

```python
#verisetimizin 150 satır 4 parametreden oluştuğunu görmekteyiz
X.shape
```

```python
y.shape
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)  
```

```python
pca.components_
```

```python
Z =pca.transform(X)
```

```python
Z.shape
```

```python
plt.scatter(Z[:,0], Z[:,1])
```

```python
plt.scatter(Z[:,0], Z[:,1], c=y)
```

```python
#bileşen sayısı ve explained varience arası ilişki grafiği
plt.figure()
plt.plot(pca.explained_variance_, 'k', linewidth=2)
plt.xlabel('n_components', fontsize=16)
plt.ylabel('explained_variance_', fontsize=16)
plt.show()
```

```python
#bileşen sayısı  olduğunda elde edilen explained varience
pca = PCA(n_components = 2) # bileşen sayısı belirlenir
pca.fit(X)

print(pca.explained_variance_ratio_)
```
